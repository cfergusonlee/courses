# Eclipse IDE Tutorial

### Workspace
Your workspace is where you store all of your projects. You shouldn't be changing its location. 

<img src='../images/Screen Shot 2019-04-24 at 3.09.59 PM.png' />

### Starting a Project

**Package:** Packages are namespaces which contain multiple packages and modules themselves. They are simply directories, but with a twist. Each package in Python is a directory which MUST contain a special file called __init__.py. This file can be empty, and it indicates that the directory it contains is a Python package, so it can be imported the same way a module can be imported.

**Module:** a piece of software that has a specific functionality. For example, when building a ping pong game, one module would be responsible for the game logic, and another module would be responsible for drawing the game on the screen. Each module is a different file, which can be edited separately.

Each project within the workspace directory gets its own directory. Eclipse organizes everything into project folders.

<img src='../images/Screen Shot 2019-04-24 at 3.13.05 PM.png' />

Eclipse creates a `.project` and `.pydevproject` file when creating a project. These are booking files that store information about the project configuration and project path, respectively.

<img src='../images/Screen Shot 2019-04-24 at 4.37.23 PM.png' />

### Perspectives

**Perspecive:** a predefined layout of views and editors. For example, the PyDev perspective is optimized for code editing while the Debugger perspective is optimized for debugging.

**View:** A view is typically used to work on a set of data. This data might be a hierarchical structure. If data is changed via the view, the underlying data is directly changed, without the need to save.

**Editor:** Editors are typically used to modify a single data element, e.g., the content of a file or a data object. Change in an editor are only applied once the user saves.

### Adding existing source code to a project

To add existing code to your project, you can just drag and drop. Choosing copy is probably a safer and more intuitive option because it leaves the original intact and you don't have to worry about remembering where all of the linked modules live.
